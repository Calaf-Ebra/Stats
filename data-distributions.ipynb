Confidence Interval ara important in checking to which extent an estimate is accuarte about a distribution or sample of data
to know if a larger sample is needed for more accuracy

```python


from scipy import stats

# Sample data (heights in inches)
heights = np.array([64, 65, 62, 68, 67, 66, 63, 64, 65, 66, 68, 69, 70, 63, 65, 66, 64, 62, 68, 67,
                    66, 65, 66, 68, 69, 70, 71, 64, 63, 65, 66, 68, 70, 67, 64, 63, 66, 68, 69, 67,
                    66, 65, 64, 63, 62, 65, 68, 70, 67, 66])

# Sample statistics
sample_mean = np.mean(heights)
sample_std = np.std(heights, ddof=1)  # ddof=1 for sample standard deviation
sample_size = len(heights)

# Calculate confidence interval (assuming 95% confidence level)
confidence_level = 0.95
alpha = 1 - confidence_level
z_critical = stats.t.ppf(1 - alpha/2, df=sample_size-1)
margin_of_error = z_critical * (sample_std / np.sqrt(sample_size))
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Print the results
print("Sample mean:", sample_mean)
print("Sample size:", sample_size)
print("Sample standard deviation:", sample_std)
print("Confidence interval (95%):", confidence_interval)

#if the interval is narrow we are good , otherwise we need larger sample size
```

A QQ Plot is a good way to check if a data sample will follow a normal path when put on a normal distribution

```python
import matplotlib.pyplot as plt
from scipy import stats
sample_data = np.array([1.2, 2.5, 3.7, 4.8, 5.2, 6.5, 7.1, 8.6, 9.9, 10.5])

# Create QQ plot
stats.probplot(sample_data, dist="norm", plot=plt)

# Add reference line
plt.plot([-3, 3], [-3, 3], color='red', linestyle='--')

# Set plot labels
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Sample Quantiles")
plt.title("QQ Plot")

# Display the plot
plt.show()
```

Bionominal Distribition helps in understanding a binary expectation given previous trials.. e.g if 20 clicks bring a sale, how many non sales will a 200 clicks bring

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
success_rate = 0.1  # Success rate of the marketing campaign
sample_size = 100  # Number of customers in the sample

# Calculate the probabilities using binomial distribution
x = np.arange(0, sample_size + 1)  # Possible number of successful conversions
probabilities = binom.pmf(x, sample_size, success_rate)

# Plot the probabilities
plt.bar(x, probabilities)
plt.xlabel("Number of Successful Conversions")
plt.ylabel("Probability")
plt.title("Binomial Distribution - Marketing Campaign")
plt.show()
```

Chi square is a good method to check if  a variable effects how another falls into a catogery..e.g Gender and salary


```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Create a sample DataFrame with gender and promotion status
data = pd.DataFrame({
    'Gender': ['Male', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Female', 'Male', 'Female'],
    'Promotion': ['Promoted', 'Not Promoted', 'Not Promoted', 'Not Promoted', 'Promoted',
                  'Not Promoted', 'Promoted', 'Not Promoted', 'Promoted', 'Promoted']
})

# Create a contingency table
contingency_table = pd.crosstab(data['Gender'], data['Promotion'])

# Perform the chi-square test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

# Define significance level
alpha = 0.05

# Print the results
print("Chi-square statistic:", chi2)
print("p-value:", p_value)
print("Degrees of freedom:", dof)
print("Expected frequencies:\n", expected)

# Check if the p-value is less than the significance level
if p_value < alpha:
    print("There is a significant association between gender and promotion.")
else:
    print("There is no significant association between gender and promotion.")
    
#If the p-value is less than the significance level, we conclude that there is a significant association,
#suggesting that gender may have an impact on promotion rates.
```

Exponential Distribution shows how something can happen within a timespan..e.g arrival rate per minute

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for exponential distribution
arrival_rate = 0.1  # Average arrival rate of 0.1 customers per minute

# Generate inter-arrival times using exponential distribution
inter_arrival_times = np.random.exponential(scale=1/arrival_rate, size=100)

# Plot the inter-arrival times
plt.hist(inter_arrival_times, bins=10, density=True)
plt.xlabel('Inter-arrival Time')
plt.ylabel('Probability Density')
plt.title('Exponential Distribution of Inter-arrival Times')
plt.show()
```

```python
###Failure of a lightbulb
import numpy as np
import matplotlib.pyplot as plt

# Parameters for Weibull distribution
shape = 2.5    # Shape parameter
scale = 100    # Scale parameter (representing the mean or characteristic lifetime)

# Generate random values from Weibull distribution
lifetime_values = np.random.weibull(shape, size=100)

# Scale the values to represent actual lifetimes
lifetime_values *= scale

# Plot the lifetime values
plt.hist(lifetime_values, bins=10, density=True)
plt.xlabel('Lifetime')
plt.ylabel('Probability Density')
plt.title('Weibull Distribution of Light Bulb Lifetimes')
plt.show()
```
